```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/content/phishing_site_urls.csv"

df = pd.read_csv(file_path)

# Feature Engineering
df["count_at"] = df["URL"].str.count("@")
df["count_slash"] = df["URL"].str.count("//")

# Summary Statistics
summary = df.groupby("Label")[["count_at", "count_slash"]].mean()
print("\n=== Average Malicious Character Counts per Label ===")
print(summary)

# Histogram for '@'
plt.figure(figsize=(8, 5))
for label in df["Label"].unique():
    subset = df[df["Label"] == label]
    plt.hist(subset["count_at"], bins=30, alpha=0.5, label=label)

plt.xlabel("Count of '@' in URL")
plt.ylabel("Frequency")
plt.title("Distribution of '@' Character Count per Label")
plt.legend()
plt.tight_layout()
plt.show()

# Histogram for '//'
plt.figure(figsize=(8, 5))
for label in df["Label"].unique():
    subset = df[df["Label"] == label]
    plt.hist(subset["count_slash"], bins=30, alpha=0.5, label=label)

plt.xlabel("Count of '//' in URL")
plt.ylabel("Frequency")
plt.title("Distribution of '//' Character Count per Label")
plt.legend()
plt.tight_layout()
plt.show()
```
